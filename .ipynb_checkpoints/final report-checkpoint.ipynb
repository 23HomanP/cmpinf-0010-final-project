## **ü•Å The Winner: ü•áüëè** 

```python
# calculate average rankings for each neighborhood based on the sum of all three rankings
for neighborhood, rankings in shared_rankings.items():
    total_ranking = sum(rankings)
    average_ranking = total_ranking / 3
    # update the dictionary with the average ranking
    shared_rankings[neighborhood].append(average_ranking)

# sort the neighborhoods based on their average ranking
sorted_rankings = sorted(shared_rankings.items(), key=lambda x: x[1][-1])

# display the sorted rankings
print("Neighborhood Rankings (1st place to last place):")
print("\t(average ranking is calculated to 2 decimal places)")
for index, (neighborhood, rankings) in enumerate(sorted_rankings, start=1):
    print(f"{index}. {neighborhood}: {rankings[-1]:.2f}")
```
