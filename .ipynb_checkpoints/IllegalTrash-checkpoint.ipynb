<h1>Illegal Dump Sites</h1>
<p>This set of data analyzes the number of illegal trash dumpsites along with the total amount of trash in each neighborhood of Pittsburgh to determine which one is the best or in my case, which one is the worst.</p>

<img src="https://th.bing.com/th/id/OIP.UNZ5fgx_qi-Pj-3PhV20OwHaEK?rs=1&pid=ImgDetMain" alt="Trubish">

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
data = pd.read_csv("IllegalTrash.csv")
```

Filter out any data not in Pittsburgh

```python
queryMask = data['City'] == "Pittsburgh"
queryMask.head(5)
df = data[queryMask]
```

Make a subset dataframe to convert into a dictionary

```python
trash = dict()
trashDf = df[["Neighborhood", "estimated_tons"]]
trashDf = trashDf.dropna(subset="estimated_tons")
```

I wanted to use a dictionary to keep track of the total amount of trash in each neighborhood becasue I
felt like the key value pairs would make the data easy to read. It'd also be easier to make a plot using the dictionary

```python
for index, row in trashDf.iterrows():
    key = row["Neighborhood"]
    if key not in trash:
        trash[key] = row["estimated_tons"]
    else:
        trash[key] = round(float(trash.get(key)) + float(row["estimated_tons"]), 2)
print(trash)
```

I do have to convert my dictionary into a dataframe so that I can make a bar graph.

```python
totalTrashDf= pd.DataFrame.from_dict(trash, orient="index").reset_index()
totalTrashDf.columns = ["Neighborhood", "Trash in Tons"]
```

```python
totalTrashDf["Trash in Tons"] = pd.to_numeric(totalTrashDf["Trash in Tons"])

totalTrashPlot =totalTrashDf.plot.bar(x="Neighborhood", y="Trash in Tons",figsize=(10,7))
print(totalTrashPlot)
```

Now to find how many illegal dump sites there are in each neighborhood!

```python
siteFq = dict()
for index, row in trashDf.iterrows():
    key = row["Neighborhood"]
    if key not in siteFq:
        siteFq[key] = 1
    else:
        siteFq[key] += 1
```

Once again I have to convert my dictionary into a dataframe so that I could make a bar graph. 

```python
print(len(siteFq))
SiteDf= pd.DataFrame.from_dict(siteFq, orient="index").reset_index()
SiteDf.columns = ["Neighborhood", "Number of Sites"]

SiteDf["Number of Sites"] = pd.to_numeric(SiteDf["Number of Sites"])

SiteDf.plot.bar(x="Neighborhood", y="Number of Sites", figsize=(10,7))
```

<h2>Here's an ordered list</h2>

```python
sitedf = SiteDf.sort_values(by="Number of Sites").reset_index()
with pd.option_context('display.max_rows', None,'display.max_columns', None,'display.precision', 3,):
    print(sitedf)
```
